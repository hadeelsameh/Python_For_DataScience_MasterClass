```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('train.csv')
```

```python
sns.pairplot(df,palette = 'coolwarm')
```

```python
sns.distplot(df['Age'],kde=False,bins =100)
```

```python
sns.jointplot(x = 'Age',y = 'Survived', data = df ,kind ='scatter')
```

```python
dic={'female':1 ,'male':0 }
df['Sex']=df['Sex'].map(dic)
sns.jointplot(x = 'Sex',y = 'Survived', data = df ,kind ='kde')
```

```python
 sns.barplot(x ='Pclass',y='Survived', data = df , estimator =np.std)
```

```python
sns.heatmap(df.corr(),cmap='magma')
```
