```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn import tree
from sklearn.model_selection import cross_val_score
```

```python
X = df.drop(["pm"],axis = 1)
Y = df["pm"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=5)
```

```python
Tree_regression_model = tree.DecisionTreeRegressor(random_state=5)
```

```python
Tree_regression_model.fit(X_train, Y_train)
Y_predicted_Tree=Tree_regression_model.predict(X_test)
```

```python
r2_score(Y_test,Y_predicted_Tree)
```

```python
Linear_regression_model = LinearRegression().fit(X_train, Y_train)
```

```python
Y_predicted_Linear = Linear_regression_model.predict(X_test)
```

```python
r2_score(Y_test,Y_predicted_Linear)
```
